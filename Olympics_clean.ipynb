```python
import numpy as np
import pandas as pd
```

```python
athlete_data = pd.read_csv('/content/athlete_events.csv')
athlete_data

```

```python
athlete_data['Age'].fillna(athlete_data['Age'].median(), inplace=True)
athlete_data['Height'].fillna(athlete_data['Height'].median(), inplace=True)
athlete_data['Weight'].fillna(athlete_data['Weight'].median(), inplace=True)
athlete_data
```

```python
athlete_data['Total_Medal'] = athlete_data['Medal'].apply(lambda x: 0 if pd.isna(x) else 1)
```

```python
def categorize_age(age):
  if age < 20:
     return 'Teen'
  elif 20 <= age < 35:
    return 'Adult'
  else:
    return 'Senior'

athlete_data['Age_Group'] = athlete_data['Age'].apply(categorize_age)
athlete_data
```

```python
athlete_data['Decade'] = (athlete_data['Year'] // 10) * 10
athlete_data.drop_duplicates(inplace=True)
athlete_data = athlete_data.dropna(subset=['Medal'])
print(athlete_data.head())
```

```python
athlete_data.to_csv('cleaned_olympics_data.csv', index=False)
```
